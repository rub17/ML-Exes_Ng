```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
#import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

import ipywidgets as widgets
from ipywidgets import interact, interact_manual

from scipy.optimize import fmin_bfgs
from scipy.optimize import fmin

from scipy.io import loadmat #To Read .mat files
import matplotlib.image as mpimg # 
from scipy.optimize import curve_fit

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12.0,6.0)
print("Setup Complete")
```

```python
X = loadmat('data/ex5data1.mat')['X']
y = loadmat('data/ex5data1.mat')['y']
Xtest = loadmat('data/ex5data1.mat')['Xtest']
ytest = loadmat('data/ex5data1.mat')['ytest']
Xval = loadmat('data/ex5data1.mat')['Xval']
yval = loadmat('data/ex5data1.mat')['yval']
X = np.squeeze(X)
y = np.squeeze(y)
Xtest = np.squeeze(Xtest)
ytest = np.squeeze(ytest)
Xval = np.squeeze(Xval)
yval = np.squeeze(yval)
train_size = len(y)
```

```python
fig, ax = plt.subplots()
ax.plot(X,y,'x')
```

## *Use Scipy curve_fit to perform the fit.

```python
def linearFunc(x_data, k,b):
    return k*x_data + b
popt,pcov = curve_fit(linearFunc, X, y)
plt.plot(X,y,'x',X,linearFunc(X,*popt),'-')
```

### Using numpy polyfit

```python
pfit,residuals,rank,singular_values,rcond = np.polyfit(X,y,1,full=True)
print("chi2 values: " + str(residuals))
plt.plot(X,y,'x',X,linearFunc(X,*pfit),'-')
```

```python
def RegCostFun(theta,m,x,y,lamb):
    J = 1/2/m*np.sum((x[:m,:]@theta-y[:m])**2) + lamb/2/m*np.sum(theta**2)
    return J
lin_Theta = np.ones((2,1))
lin_X = np.hstack((np.ones((train_size,1)),X.reshape(train_size,1)))
lin_y = y.reshape((train_size,1))
RegCostFun(lin_Theta,train_size,lin_X,lin_y,1)
```

```python
def RegCostGrad(theta,m,x,y,lamb):
    gJ0 = 1/m*(np.sum((x[:m,0].reshape(m,-1)@theta[0].reshape(1,-1)-y[:m])*x[:m,0].reshape(-1,1)))
    theta_i = theta[1:]
    x_i = x[:m,1:]
    gJi = 1/m*(np.sum((x_i@theta_i-y[:m])*x_i)) + lamb/m*theta_i
    return gJ0,gJi
RegCostGrad(lin_Theta,12,lin_X,lin_y,1)
```

```python
trd_Theta = fmin_bfgs(RegCostFun,lin_Theta,args=(train_size,lin_X,y,0))
plt.plot(X,y,'x',X,lin_X@trd_Theta,'-')
```

# Bias and Variance

```python
def learningCurve(theta,size,x,y,xval,yval,lamb):
    Theta_History = []
    Train_err = []
    Val_err = [] 
    for i in range(size):
        thta = fmin_bfgs(RegCostFun,theta,args=(i+1,x,y,lamb),disp=False)
        Theta_History.append(thta)
        Train_err.append(RegCostFun(thta,i+1,x,y,lamb))
        Val_err.append(RegCostFun(thta,size,xval,yval,lamb))
    return Train_err, Val_err

lin_Xval = np.hstack((np.ones((train_size,1)),Xval[0:train_size].reshape(train_size,1)))
lin_yval = yval[:train_size]
Train_err,Val_err= learningCurve(lin_Theta,11,lin_X,y,lin_Xval,lin_yval,0)
m = np.linspace(2,12,11)
plt.plot(m,Train_err,'-',m,Val_err,'-')
plt.ylim(0,150)
```

```python
def polyFeatures(x,deg):
    x_out = x.reshape(-1,1)
    for i in range(2,deg+1):
        higher = (x**i).reshape(-1,1)
        x_out = np.hstack((x_out,higher))
    return x_out
def featureNorm(x):
    mean = np.mean(x,axis=0)
    std = np.std(x,axis=0)
    return (x-mean)/std, mean, std
```

```python
X_norm,X_mean,X_std = featureNorm(X)
X_8 = polyFeatures(X_norm,8)
X_8 = np.hstack((np.ones((len(X),1)),X_8))
Theta_8 = np.ones((9,1))
trd_Theta8 = fmin_bfgs(RegCostFun,Theta_8,args=(train_size,X_8,y,0))
sorted_index = np.argsort(X) ### Need sorting to prevent messy line plots.
y_8 = X_8@trd_Theta8
plt.plot(X[sorted_index],y[sorted_index],'x',X[sorted_index],y_8[sorted_index],'--')
```

```python
Xval_8 = polyFeatures(featureNorm(Xval)[0],8)
Xval_8 = np.hstack((np.ones((12,1)),Xval_8[:12,:]))
Train_8_err,Val_8_err = learningCurve(Theta_8,11,X_8,y,Xval_8,yval,0)
plt.plot(m,Train_8_err,'-',m,Val_8_err,'-')
plt.ylim(0,100)
```
