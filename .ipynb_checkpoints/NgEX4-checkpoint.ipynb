## Feedforward and regularized cost function

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
def sigmoidgrad(z):
    return sigmoid(z)*(1-sigmoid(z))
def nnRegCstFun(nnparams,input_size,hidden_size,sample_size,num_labels,
             x,y,lamb):
    theta1 = nnparams[0:(input_size+1)*hidden_size].reshape(hidden_size,input_size+1) #+1 for bias node.
    theta2 = nnparams[(input_size+1)*hidden_size:].reshape(num_labels,hidden_size+1) #NEED TO FOLLOW THE ORIGINAL WAY TO UNRAVEL!!!!!!!!! DONT SKIP AHEAD.
    z2 = x@theta1.T
    a2 = np.hstack((np.ones((sample_size,1)),sigmoid(z2)))#add a bias node.
    z3 = a2@theta2.T
    h = sigmoid(z3)
    J = 0

    for i in range(sample_size):
        yi = np.zeros((10,1))
        a = y[i][0]
        yi[a-1]=1 #THIS IS STUPID... MISALIGNMENT FROM MATLAB DATA TO PYTHON.
        #reg_term = (lambda_ / (2 * m)) * (np.sum(np.square(temp1[:, 1:])) + np.sum(np.square(temp2[:, 1:])))
        reg_term = lamb/sample_size/2*(np.sum(np.square(theta1[:,1:])) + np.sum(np.square(theta2[:,1:]))) #Skip the bias node.
        
    J += (-np.log(h[i]).dot(yi)-np.log(1-h[i]).dot((1-yi))) + reg_term
    return J.item()/sample_size #return the value only
```

## Backward Propagation

```python
def rand_init_weights(Lin,Lout):
    epsilon_init = 6**(0.5)/ (Lin + Lout)**(0.5) ## general good choice for epsilon value depends on the # of nodes.
    w = np.random.rand(Lin+1,Lout)*2*epsilon_init-epsilon_init
    return w
    
init_Theta1 = rand_init_weights(400,25) #401,25
init_Theta2 = rand_init_weights(25,10) #26,10
init_Theta = np.append(init_Theta1.flatten(),init_Theta2.flatten())
```

```python
def nnRegCostBack(nnparams,input_size,hidden_size,num_labels,sample_size,x,y,lamb):
    grad1=grad2=0
    theta1 = nnparams[0:(input_size+1)*hidden_size].reshape(hidden_size,input_size+1) #+1 for bias node.
    theta2 = nnparams[(input_size+1)*hidden_size:].reshape(num_labels,hidden_size+1) #NEED TO FOLLOW THE ORIGINAL WAY TO UNRAVEL!!!!!!!!! DONT SKIP AHEAD.
    for t in range(sample_size):
        a1 = Xin[t] #401,
        z2 = a1.dot(init_Theta1) #25,
        a2 = np.append(1,sigmoid(z2)) #26
        z3 = a2.dot(init_Theta2) #10,
        yt = np.zeros(10)
        yt[y[t][0]-1] = 1
        err3 = yt-sigmoid(z3) #10,
        err2 = init_Theta2.dot(err3)[1:]*sigmoidgrad(z2) #25
        grad2 += err3.reshape(-1,1)@(a2.reshape(1,-1))
        grad1 += err2.reshape(-1,1)@(a1.reshape(1,-1))
    grad1 = grad1/sample_size
    grad2 = grad2/sample_size
    temp1 = np.hstack((np.zeros((theta1.shape[0],1)),theta1[:,1:]))
    temp2 = np.hstack((np.zeros((theta2.shape[0],1)),theta2[:,1:]))
    grad1_reg = grad1 + lamb/sample_size * temp1
    grad2_reg = grad2 + lamb/sample_size * temp2
    grad_vec = np.append(grad1_reg.flatten(), grad2_reg.flatten())
    return grad_vec
```

```python
itr = 200
alpha = 0.01
def grdDsnt(alpha,itr,nnparams,input_size, hidden_size, num_labels,x,y,lamb):
    J_hist = []
    m = len(y)
    theta1 = theta2 = 0
    for i in range(itr):
        gradVec = nnRegCostBack(nnparams,input_size,hidden_size,num_labels,m,x,y,lamb)
        theta1 -= alpha*(gradVec[0:(input_size+1)*hidden_size].reshape(hidden_size,input_size+1))
        theta2 -= alpha*(gradVec[(input_size+1)*hidden_size:].reshape(num_labels,hidden_size+1))
        new_theta = np.append(theta1.flatten(),theta2.flatten())
        J = nnRegCstFun(new_theta,input_size,hidden_size,m,num_labels,x,y,lamb)
        J_hist.append(J)
    thetaVec = np.append(theta1.flatten(),theta2.flatten())
    return thetaVec,J_hist

theta_after, J_history = grdDsnt(alpha,itr,init_Theta,400,25,10,Xin,y,1)
```
